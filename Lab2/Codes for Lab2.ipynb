```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
import scipy.optimize as optim
from matplotlib.pyplot import MultipleLocator
```

```python
rp=np.array([1,
0.95,
0.90,
0.40,
0.22])
amp=np.array([1,
0.94,
0.82,
0.29,
0.14])
amp_2=np.array([1,
0.94,
0.82,
0.29,
0.14])
for i in range(5):
    amp_2[i]=amp[i]**2
f_1=np.array([10,50,100,500,1000])
log=np.array([1,1.7,2,2.7,3])
```

```python
from scipy import signal
```

```python
s1 = signal.TransferFunction([0,0,1], [0,0.00628,1])
f, mag, phase = s1.bode()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(18, 10))
ax2=ax.twinx()
ax2.set_ylim(-85,0)
ax.set_ylim(-20,1)
ax.semilogx(f, mag,label='theoretical gain')
ax2.semilogx(f, phase,'--',c='r',label='theoretical phase')
ax.legend(loc=0,numpoints=1,fontsize=15)
ax2.legend(loc=2,bbox_to_anchor=(0.8,0.93),numpoints=1,fontsize=15)
ax.tick_params(axis='both',labelsize=20)
ax2.tick_params(axis='both',labelsize=20)
ax.set_ylabel('Gain(dB)',fontsize=24)
ax2.set_ylabel("Phase($^{\circ}$)",fontsize=24)
ax.set_xlabel("Frequency(Hz)",fontsize=24)
ax.grid(True,which="minor")
ax.grid(True,which="major")
ax2.grid(True,which="minor")
ax2.grid(True,which="major")
plt.show()
```
