```python
import numpy as np
import scipy.stats as st
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
from sklearn import linear_model,metrics
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.optimize as optim
import math

df1 = pd.read_csv('LTspice Data.csv')
```

```python
df1.head
```

```python
V_diode=np.zeros(1000)
I_diode=np.zeros(1000)
V_temp=df1[['V(vdiode)']].values
I_temp=df1[['I(D1)']].values
for i in range(1000):
    V_diode[i]=V_temp[i]
    I_diode[i]=I_temp[i]
```

```python
def i_diode(v,Is,Vt):
    return Is*(math.e**(v/Vt))-Is
```

```python
x1=np.linspace(0,0.63,1000)
```

```python
bnd = (0, [10,10])
(A, B), cov = optim.curve_fit(i_diode, V_diode, I_diode, bounds=bnd)
print(A,B,cov)
y_lrp=i_diode(x1,A,B)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(18, 10))
ax.plot(V_diode*1000, I_diode*1000000,label='measured points',linewidth=10)
ax.plot(x1*1000,y_lrp*1000000,'-',label='best fit line',linewidth=3)
ax.set_xlim(0, 700)
ax.set_ylim(0, 400)
plt.xlabel("${V_d}{_i}{_o}{_d}{_e}$(mV)",fontsize=24)
plt.ylabel("${I_d}{_i}{_o}{_d}{_e}$($\mu$A)",fontsize=24)
plt.legend(loc=0,numpoints=1)
leg=plt.gca().get_legend()
ltext=leg.get_texts()
plt.setp(ltext,fontsize='xx-large')
plt.tick_params(axis='both',labelsize=20)
plt.grid()
```
