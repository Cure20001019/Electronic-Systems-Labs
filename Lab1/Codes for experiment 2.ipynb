```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
import scipy.optimize as optim
from matplotlib.pyplot import MultipleLocator
```

```python
rl=np.array([4.7,10,20,47,70,100])
vol=np.array([0.84,1.36,1.88,2.39,2.56,2.68])
```

```python
def LogisticGrowth(t, a, b, c):
    return c / (1+a*np.exp(-b*t))
```

```python
def TheoryVout(x):
    return 30*x/[20*(30+x)+30*x]*5
```

```python
x1=np.linspace(6,65,100)
x2=np.linspace(0,120)
y=TheoryVout(x1)
y=y.reshape(-1,1)
```

```python
bnd = (0, [10,10,10])
(A, B,C), cov = optim.curve_fit(LogisticGrowth, rl, vol, bounds=bnd)
print(A,B,C)
y_pred=LogisticGrowth(x2,A,B,C)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(18, 10))
ax.plot(rl, vol, 'ok',ms=10,label='measured points')
ax.plot(x1,y,'g',label='theory',linewidth=3.5)
ax.plot(x2,y_pred,'--',label='best fit line',linewidth=3.5)
ax.set_xlim(0, 120)
ax.set_ylim(0.5, 3)
plt.ylabel("${V_o}{_u}{_t}$(V)",fontsize=24)
plt.xlabel("${R_L}$(k$\Omega$)",fontsize=24)
plt.legend(loc=0,numpoints=1)
leg=plt.gca().get_legend()
ltext=leg.get_texts()
plt.setp(ltext,fontsize='xx-large')
plt.tick_params(axis='both',labelsize=20)
plt.grid()
plt.show()
```
