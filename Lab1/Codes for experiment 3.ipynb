```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
import scipy.optimize as optim
from matplotlib.pyplot import MultipleLocator
```

```python
v=np.array([0.75,
1.20,
1.71,
2.27,
2.61])
i=np.array([-150.00,
-120.00,
-85.71,
-48.39,
-26.09])
```

```python
def Theory(x):
    return  0.015*x+3
```

```python
x1=np.linspace(-160,0,100)
x2=np.linspace(-110,-30)
x3=np.linspace(-160,0,100)
y=Theory(x2)
y=y.reshape(-1,1)
```

```python
lr = lm.LinearRegression()
lr.fit(i[:, np.newaxis], v)
y_lr = lr.predict(x1[:, np.newaxis])
y_lr2 = lr.predict(x3[:, np.newaxis])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(18, 10))
ax.plot(i, v, 'ok',ms=10,label='measured points')
#ax.plot(x1,y_lr,'r','--',label='best fit for Thévenin equivalent',linewidth=3.5)
ax.plot(x1,y_lr,'--',label='best fit line',linewidth=3.5)
ax.set_xlim(-160, 0)
ax.set_ylim(0.5, 3.5)
plt.ylabel("${V_o}{_u}{_t}$(V)",fontsize=24)
plt.xlabel("${I_o}{_u}{_t}$($\mu$A)",fontsize=24)
plt.legend(loc=0,numpoints=1)
leg=plt.gca().get_legend()
ltext=leg.get_texts()
plt.setp(ltext,fontsize='xx-large')
plt.tick_params(axis='both',labelsize=20)
plt.grid()
```
